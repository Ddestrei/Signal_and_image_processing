```python
import skimage
import matplotlib.pyplot as plt
import numpy as np

percent = 10
img = np.zeros((128, 128), np.uint8) + 127

noise = np.random.uniform(low=0, high=255, size=(128, 128))
gausse = np.random.normal(loc=0, scale=1, size=(128, 128))

plt.hist(noise.ravel(), bins=256)
plt.show()
plt.hist(gausse.ravel(), bins=256)
plt.show()
noise_0 = noise < 10
noise_1 = noise > 245
#img = skimage.util.random_noise(img, 's&p',amount=percent/100)

img[noise_1] = 255
img[noise_0] = 0

plt.imshow(img, cmap='gray')

f = np.random.uniform(low=0, high=255, size=(128, 128))
e = np.random.uniform(low=0, high=255, size=(128, 128))

denominator = 0.0

sum_over = 0.0
for x in range(f.shape[0]):
    for y in range(f.shape[1]):
        #for c in range(f.shape[2])
        sum_over += (f[x, y] - e[x, y]) ** 2
        denominator += f[x, y] ** 2

print(sum_over / denominator)
```
