## Macierze numpy
### Szachownica
Przy pomocy macierzy numpy stwórz szachownicę o wymiarach AxA. Pola czarne powinny mieć wartość 0, a białe 255.

https://numpy.org/doc/stable/user/basics.indexing.html

```python
import matplotlib.pyplot as plt
import numpy as np
import time


A = 2048
array = np.empty((A, A), dtype=np.uint8)

start = time.time()

# Umieść swoje rozwiązanie tutaj

SIDE_LENGTH = int(A/8)
c = 0
for i in range(8):
    for j in range(8):
        if c % 2 == 0:
            array[i*SIDE_LENGTH:i*SIDE_LENGTH+SIDE_LENGTH-1,j*SIDE_LENGTH:j*SIDE_LENGTH+SIDE_LENGTH-1] = 255
        else:
            array[i*SIDE_LENGTH:i*SIDE_LENGTH+SIDE_LENGTH-1,j*SIDE_LENGTH:j*SIDE_LENGTH+SIDE_LENGTH-1] = 0
        c+=1
    c+=1
        
print("Elapsed time: ", time.time() - start)
plt.imshow(array, cmap='gray', interpolation='nearest')
plt.show()
```

## Plus

Przy pomocy macierzy numpy stwórz plus o wymiarach AxA. Środek powinien być tłem. Postaraj się korzystać jedynie z indeksowania macierzami numpy.

```python
import matplotlib.pyplot as plt
import numpy as np
import time


A = 2048
thicknes = 256
array = np.empty((A, A), dtype=np.uint8)
start = time.time()

array[0:A,0:A] = 255

array[824:1223,0:823] = 0
array[824:1223,1224:2047] = 0
array[0:823,824:1223] = 0
array[1224:2047,824:1223 ]=0
# Umieść swoje rozwiązanie tutaj

print("Elapsed time: ", time.time() - start)
plt.imshow(array, cmap='gray', interpolation='nearest')
plt.show()
```

## Obrazy

Wczytaj obrazy (skimage/cv2) z dane/image_1.png i dane/image_2.png. Następnie wyświetl je obok siebie tak jak na poniższym obrazku. Pamiętaj o dodaniu tytułów do obrazów.

Obraz 3 powinien być sumą obrazów 1 i 2 podzieloną przez 2.

Wiersz pierwszy powinien zawierać obrazy RGB, wiersz drugi powinien zawierać te same obrazy, ale w skali szarości.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html#examples-using-matplotlib-pyplot-imshow

```python
import cv2
import skimage
import matplotlib.pyplot as plt

# Umieść swoje rozwiązanie tutaj

image1 = cv2.imread('dane/image_1.png', cv2.IMREAD_COLOR_RGB)
image2 = cv2.imread('dane/image_2.png', cv2.IMREAD_COLOR_RGB)
image3 = cv2.addWeighted(image1, 0.5, image2,0.5,0)
fig, axes = plt.subplots(2,3,figsize=(15, 7))
axes.flat[0].imshow(image1)
axes.flat[1].imshow(image2)
axes.flat[2].imshow(image3)
axes.flat[3].imshow(cv2.cvtColor(image1, cv2.COLOR_RGB2GRAY),cmap='gray')
axes.flat[4].imshow(cv2.cvtColor(image2, cv2.COLOR_RGB2GRAY),cmap='gray')
axes.flat[5].imshow(cv2.cvtColor(image3, cv2.COLOR_RGB2GRAY),cmap='gray')
plt.show()
```
