```python
import numpy as np
import matplotlib.pyplot as plt

matrix = np.array([[0.5, 0.5, 0.5,0.5,0.5], [0.5, 1, 0.5,0.5,0.5], [0.5, 1, 0.5,0.5,0.5],[0.5, 0.5, 0.5,0.5,0.5],[0.5, 0.5, 0.5,0.5,0.5],[0.5, 0.5, 0.5,0.5,0.5],
                   [0.5, 0.5, 0.5,0.5,0.5],[0.5, 0.5, 0.5,0.5,0.5],[0.5, 0.5, 0.5,0.5,0.5]])
print(matrix.shape)
plt.imshow(matrix, cmap='gray')
```

```python
def horizontal_mirror(matrix):
    return matrix[::-1,...]
horizontal_mirror = horizontal_mirror(matrix)
plt.imshow(horizontal_mirror, cmap='gray')
```

```python
def vertical_mirror(matrix):
    return matrix[...,::-1]
vertical_mirror = vertical_mirror(matrix)
plt.imshow(vertical_mirror, cmap='gray')
```

```python
def right_turn(matrix):
    return np.rot90(matrix,3)
right_turn = right_turn(matrix)
plt.imshow(right_turn, cmap='gray')
```

```python
def left_turn(matrix):
    return np.rot90(matrix)
left_turn = left_turn(matrix)
plt.imshow(left_turn, cmap='gray')
```

```python
def half_turn(matrix):
    return np.rot90(matrix,2)
half_turn = half_turn(matrix)
plt.imshow(half_turn, cmap='gray')
```

```python
def expanded_matrix(matrix):
    matrix_x, matrix_y = matrix.shape
    if matrix_x > matrix_y:
        expanded_matrix = np.zeros((matrix_x, matrix_x))
        expanded_matrix[:,int((matrix_x-matrix_y)/2):int(matrix_x-(matrix_x-matrix_y)/2)] = matrix
        return expanded_matrix
    return None
expanded_matrix = expanded_matrix(matrix)
plt.imshow(expanded_matrix, cmap='gray')
print(expanded_matrix)
```

```python

def reduced_matrix(matrix):
    matrix_x, matrix_y = matrix.shape
    if matrix_x > matrix_y:
        reduced_matrix = np.zeros((matrix_y, matrix_y))
        reduced_matrix = matrix[int((matrix_x-matrix_y)/2):int(matrix_x-(matrix_x-matrix_y)/2),:]
        return reduced_matrix
    return None
reduced_matrix = reduced_matrix(matrix)
plt.imshow(reduced_matrix, cmap='gray')
print(reduced_matrix)
```

```python
from random import randint, shuffle

dog = cv2.imread('dog.jpeg', cv2.IMREAD_GRAYSCALE)
matrix_x, matrix_y = dog.shape
if matrix_x > matrix_y:
    dog = dog[int((matrix_x-matrix_y)/2):int(matrix_x-(matrix_x-matrix_y)/2),:]
else:
    dog = dog[:,int((matrix_y-matrix_x)/2):int(matrix_y-(matrix_y-matrix_x)/2)]
plt.imshow(dog, cmap='gray')
```

```python
N = 5
LENGTH = 134
X = np.array([0,1,2,3,4])
shuffle(X)
X = list(X)
new_dog = np.zeros(dog.shape)
for i in range(N):
    x = X.pop()
    Y = np.array([0,1,2,3,4])
    shuffle(Y)
    Y = list(Y)
    for j in range(N):
        y = Y.pop()
        part = dog[i*LENGTH:i*LENGTH+LENGTH, j*LENGTH:j*LENGTH+LENGTH]
        new_dog[x*LENGTH:x*LENGTH+LENGTH, y*LENGTH:y*LENGTH+LENGTH] = part

plt.imshow(new_dog, cmap='gray')
```
