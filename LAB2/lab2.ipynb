```python
import matplotlib.pyplot as plt
import numpy as np
import skimage as sk

image = sk.data.camera()
image.dtype

plt.subplots(2, 2, figsize=(12, 12))
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.subplot(2, 2, 2)
hist = np.histogram(image.flatten(), bins=256)[0]
plt.plot(hist)
bar = np.histogram(image.flatten() // 32, bins=8)[0]
plt.subplot(2, 2, 3)
plt.bar(np.arange(256 // 32), bar)
plt.subplot(2, 2, 4)
cumulative = np.cumsum(hist)
plt.plot(cumulative)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import skimage as sk

image = sk.data.camera()
plt.subplots(2, 2, figsize=(12, 12))
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.subplot(2, 2, 3)
hist = np.histogram(image.flatten(), bins=256)[0]
plt.plot(hist)
plt.subplot(2, 2, 2)
cumulative = sk.exposure.rescale_intensity(image, in_range=(5, 250))
plt.imshow(cumulative, cmap='gray')
plt.subplot(2, 2, 4)
hist1 = np.histogram(cumulative, bins=256)[0]
plt.plot(hist1)
plt.show()
unique_vals = np.unique(cumulative)
print(len(unique_vals), "unikalnych wartości")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import skimage as sk

chelsea = sk.data.chelsea()
red = np.zeros(chelsea.shape[0] * chelsea.shape[1], dtype=np.uint8)
green = np.zeros(chelsea.shape[0] * chelsea.shape[1], dtype=np.uint8)
blue = np.zeros(chelsea.shape[0] * chelsea.shape[1], dtype=np.uint8)
counter = 0
for i in range(chelsea.shape[0]):
    for j in range(chelsea.shape[1]):
        red[counter] = chelsea[i, j][0]
        green[counter] = chelsea[i, j][1]
        blue[counter] = chelsea[i, j][2]
        counter += 1

red_h = np.histogram(red, bins=256)[0]
green_h = np.histogram(green, bins=256)[0]
blue_h = np.histogram(blue, bins=256)[0]
red_h = [i for i in red_h if i !=0]
green_h = [i for i in green_h if i !=0]
blue_h = [i for i in blue_h if i !=0]
plt.subplots(2, 1, figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(red_h, label='red', color='r')
plt.plot(green_h, label='green', color='g')
plt.plot(blue_h, label='blue', color='b')
plt.subplot(2, 1, 2)
plt.bar(np.arange(8)-0.25,np.histogram(red//32, bins=256//32)[0], label='red', color='r',width=0.25)
plt.bar(np.arange(8),np.histogram(green//32, bins=256//32)[0], label='green', color='g',width=0.25)
plt.bar(np.arange(8)+0.25,np.histogram(blue//32, bins=256//32)[0], label='blue', color='b',width=0.25)
plt.show()
```
