```python
import matplotlib.pyplot as plt
import numpy as np
import skimage as sk

image = sk.data.camera()
image.dtype

plt.subplots(2,2,figsize=(12,12))
plt.subplot(2,2,1)
plt.imshow(image, cmap='gray')
plt.subplot(2,2,2)
hist = np.histogram(image.flatten(), bins=256)[0]
plt.plot(hist)
bar = np.histogram(image.flatten()//32, bins=8)[0]
plt.subplot(2,2,3)
plt.bar(np.arange(256//32),bar)
plt.subplot(2,2,4)
cumulative = np.cumsum(hist)
plt.plot(cumulative)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import skimage as sk

image = sk.data.camera()
plt.subplots(2,2,figsize=(12,12))
plt.subplot(2,2,1)
plt.imshow(image, cmap='gray')
plt.subplot(2,2,3)
hist = np.histogram(image.flatten(), bins=256)[0]
plt.plot(hist)
plt.subplot(2,2,2)
cumulative = sk.exposure.rescale_intensity(image, in_range=(5, 250))
plt.imshow(cumulative,cmap='gray')
plt.subplot(2,2,4)
hist1 = np.histogram(cumulative, bins=256)[0]
plt.plot(hist1)
plt.show()
unique_vals = np.unique(cumulative)
print(len(unique_vals), "unikalnych wartoci")
```
